# **Hypothesis Testing**

**Question 1**: What is a null hypothesis (H₀) and why is it important in hypothesis testing?

**solution:**

**Definition**

The null hypothesis is a default statement or assumption that there is no significant difference, no effect, or no relationship between the variables being studied. It typically proposes that any observed difference in data is due to random chance or sampling error rather than a specific cause.

**Why It Is Important**

The null hypothesis is crucial to the scientific method for several reasons:

Baseline for Comparison: It provides a clear, numerical benchmark (e.g., $\mu = 10$) against which the sample data is measured.

The Principle of Falsification: In statistics, we do not aim to "prove" a researcher's theory is true; instead, we attempt to gather enough evidence to reject the null hypothesis. If the evidence is strong enough, we move toward the alternative hypothesis ($H_{a}$).

Defining Error Rates: It allows researchers to set a significance level ($\alpha$), which is the probability of accidentally rejecting the null hypothesis when it is actually true (a Type I error).

Objectivity: By starting with the assumption that "nothing is happening," researchers maintain a neutral stance, ensuring that claims of "new discoveries" are backed by rigorous mathematical evidence.



**Question 2:** What does the significance level (α) represent in hypothesis testing?

**solution:**

In hypothesis testing, the significance level, denoted by the Greek letter alpha ($\alpha$), represents the probability of making a Type I error.

Key Characteristics

Threshold for Rejection: It is the predefined threshold used to decide whether to reject the null hypothesis ($H_0$).

Probability of "False Positives": It represents the specific risk a researcher is willing to take of rejecting the null hypothesis when it is actually true.

Common Values: In the provided exercises, the significance level is set at $\alpha = 0.05$. This means there is a 5% risk of concluding that a difference exists when there is actually no real difference.

Decision Rule: If the calculated p-value from a test is less than or equal to $\alpha$, the results are considered "statistically significant," and the null hypothesis is rejected.



**Question 3**: Differentiate between Type I and Type II errors.

**solution:**

**Type I Error (False Positive)**

**Definition:** This occurs when you reject the null hypothesis ($H_0$) when it is actually true.

**Probability:** The probability of committing this error is denoted by $\alpha$ (the significance level).

Example: Concluding that a training program improved productivity (rejecting $H_0$) when, in reality, the change was just due to random chance ( $H_0$ is true).

**Type II Error (False Negative)**

**Definition**: This occurs when you fail to reject the null hypothesis ($H_0$) when it is actually false.

**Probability:** The probability of committing this error is denoted by $\beta$.

Example: Concluding that a company's average resolution time is 10 minutes (failing to reject $H_0$) when it is actually significantly different ( $H_0$ is false).

**Question 4:** Explain the difference between a one-tailed and two-tailed test. Give an example of each.

**solution:**

**One-Tailed Test**

A one-tailed test is used when the research hypothesis predicts a specific direction for the effect—either an increase or a decrease—but not both.

* **Rejection Region:** The entire significance level ($\alpha$) is located in just one tail (either the far left or far right) of the distribution.

* **Example:** Testing if a new productivity training program increases employee output. The null hypothesis would be that productivity stayed the same or decreased, while the alternative specifically looks for an improvement.

**Two-Tailed Test**

A two-tailed test is used when the research hypothesis predicts that there is a difference, but does not specify the direction. It tests for both an increase and a decrease simultaneously.

**Rejection Region:** The significance level ($\alpha$) is split in half (e.g., $0.025$ in each tail for $\alpha = 0.05$), placing rejection zones at both ends of the distribution.

**Example:** Testing a company's claim that the average time to resolve a complaint is exactly 10 minutes. You would reject the claim if the actual time is significantly lower than 10 minutes or significantly higher than 10 minutes.

**Question 5**: A company claims that the average time to resolve a customer complaint is 10 minutes.

A random sample of 9 complaints gives an average time of 12 minutes and a standard deviation of 3 minutes.

At α = 0.05, test the claim.

**solution:**

To test the company's claim, we will perform a One-Sample t-test because the sample size is small ($n < 30$) and the population standard deviation is unknown.

**Step 1**: State the Hypotheses

Null Hypothesis ($H_0$): $\mu = 10$ (The average resolution time is 10 minutes).
Alternative Hypothesis ($H_a$): $\mu \neq 10$ (The average resolution time is not 10 minutes).Note: This is a two-tailed test because we are testing if the claim is "true" or "false" regardless of the direction of the difference.

**Step 2**: Identify the Data

* Claimed Mean ($\mu$): $10$

 * Sample Mean ($\bar{x}$): $12$

 * Sample Standard Deviation ($s$): $3$

  * Sample Size ($n$): $9$
  
  * Significance Level ($\alpha$): $0.05$
  
  * Degrees of Freedom ($df$): $n - 1 = 9 - 1 = 8$

**Step 3:** Calculate the t-statisticUsing the formula for the t-score:$$t = \frac{\bar{x} - \mu}{s / \sqrt{n}}$$$$t = \frac{12 - 10}{3 / \sqrt{9}} = \frac{2}{3 / 3} = \frac{2}{1} = 2.0$$

**Step 4**: Determine the Critical Value

Using a t-distribution table for $df = 8$ and a two-tailed $\alpha = 0.05$:

The critical t-value ($t_{crit}$) is approximately $\pm 2.306$.

**Step 5**: Make a Decision

Calculated t: $2.0$

Critical t-range: $-2.306$ to $+2.306$

Since $2.0$ falls within the range (it is less than $2.306$), we fail to reject the null hypothesis ($H_0$).

Conclusion: At the $0.05$ significance level, there is not enough statistical evidence to reject the company's claim that the average resolution time is 10 minutes

**Question 6**: When should you use a Z-test instead of a t-test?

**solution:**

Choosing between a Z-test and a t-test depends primarily on two factors: your sample size and whether you know the population's standard deviation.

**Use a Z-test when:**

* **Population Variance is Known:** You should use a Z-test if the population standard deviation ($\sigma$) is already known.

* **Large Sample Size**: It is appropriate to use a Z-test if your sample size is large, typically defined as $n > 30$. This is because, according to the Central Limit Theorem, the sampling distribution of the mean becomes approximately normal as the sample size increases.

**Use a t-test when:**

* **Population Variance is Unknown**: You must use a t-test if the population standard deviation is unknown and you have to estimate it using the sample standard deviation ($s$).

* **Small Sample Size:** A t-test is required when the sample size is small, specifically $n < 30$. The t-distribution has "heavier tails" than the normal distribution, which accounts for the extra uncertainty involved with smaller samples.

**Step 4: Calculate the Chi-Square Statistic ($\chi^2$)**

We use the formula $\chi^2 = \sum \frac{(O-E)^2}{E}$:

*  Male/B:  $(20 - 30)^2 / 30 = 100 / 30 \approx \mathbf{3.33}$

**Total**   $\chi^2$ $= 5 + 3.33 + 5 + 3.33 = \mathbf{16.66}$

**Step 5: Determine the Critical Value and Decision**

* Degrees of Freedom ($df$): $(Rows - 1) \times (Cols - 1) = (2-1) \times (2-1) = \mathbf{1}$.

*  Critical Value ($\chi^2_{crit}$): For $df=1$ at $\alpha=0.05$, the value is 3.841.
